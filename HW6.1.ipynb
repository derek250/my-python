```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### loss function 為 (x+1)(x+2)(x+3)(x+4)

```python
x=np.linspace(-5,0,1000)
def f(x):
    return (x+1)*(x+2)*(x+3)*(x+4)
plt.plot(x,f(x));
```

### 斜率方程為4x^3+30x^2+70x+50

```python
def df(x):
    return 4*x**3+30*x**2+70*x+50
```

### gradient descent 程式

```python
def gd(x,times,μ):
    x_start=x
    y_start=f(x)
    x_run=[]
    y_run=[]
    for i in range(0,times):
        x_run.append(x_start)
        y_run.append(f(x_start))
        x_start=x_start-(df(x_start)*μ)
        y_start=f(x_start)
    return (x_run,y_run)
```

### x隨機從-5到0之間開始迭代

```python
x_random=np.random.rand()*-5
s=gd(x_random,100,0.01)
plt.plot(x,f(x));
plt.scatter(s[0],s[1],c='r')
print(f'從x={x_random}開始以μ=0.01迭代100次')
print(f'迭代後算得x={s[0][99]}有y={s[1][99]}')
```
