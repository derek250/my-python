```python
def myOptimAction(priceVec, transFeeRate):
    import numpy as np
    import pandas as pd
    dataLen=len(priceVec)
    #Initializing MaxStockHolding and MaxCapital. These two can be seen as accumulated stockholding and capital respectively.
    MaxCapital=1
    MaxStockHolding=0
    #We set -100 as an indication of NaN.
    buyPoints=[-100]
    sellPoints=[-100]
    for i in range(dataLen):
     #For every price points, we calculate using the formula given by professor.
        stockHolding=MaxCapital*(1-transFeeRate)/priceVec[i]
        capital=MaxStockHolding*(1-transFeeRate)*priceVec[i]
       #If the stockHolding>MaxStockHolding, we collect the buy points in a list
        if stockHolding>MaxStockHolding:
            MaxStockHolding=stockHolding
            buyPoints.append(i)
       #If stockHolding<=MaxStockHolding, we retain the same buy points as before.
        elif stockHolding<MaxStockHolding or stockHolding == MaxStockHolding:
            buyPoints.append(buyPoints[i])
       #If the stockHolding>MaxStockHolding, we collect the sell points in a list
        if capital>MaxCapital:
            MaxCapital=capital
            sellPoints.append(i)
       #If stockHolding<=MaxStockHolding, we retain the same sell points as before.
        elif capital<MaxCapital or capital==MaxCapital:
            sellPoints.append(sellPoints[i])
    del sellPoints[0] #To delete the -100 
    del buyPoints[0]  #To delete the -100
  
  #We remove all -100 in the buypoints and sellpoints lists since they are NaN.
    indices=[i for i,x in enumerate(sellPoints) if x==-100] #Find if an element exists in the list
    max(indices)
    for x in range(max(indices)+1):
        sellPoints[x]=sellPoints[max(indices)+1]
    
  #Since the buyPoints and sellPoints are list the 
  #tell us the best points to buy and sell but it keeps updating and there are 
  #repetition of buy points. Hence this backtracking helps to determine 
  #the "local" most updated buy points and sell points. 
    actionVec = np.zeros(dataLen)
    actionVec=list(actionVec)
    x=1
    temp=sellPoints[-1]
    actionVec[temp]=-1
    counter=0
    buy=[0]
    sell=[temp]
    counter2=0
    while (1): 
        if x==-1:
            temp=sellPoints[temp]
            if temp==sell[counter2]:
                break
            else:
                sell.append(temp)
                x=-x
                counter2+=1
        elif x==1:
            temp=buyPoints[temp] 
            if temp==buy[counter]:
                break
            else: 
                buy.append(temp)
                x=-x
                counter+=1
    del buy[0]
    for s,b in zip(sell,buy):
        actionVec[s]=-1
        actionVec[b]=1
    return actionVec
```

```python
def profitEstimateOpen(priceVec, transFeeRate, actionVec):
    import numpy as np
    import pandas as pd
    
    capital=1
    capitalOrig=capital
    dataCount=len(priceVec)
    suggestedAction=actionVec
    stockHolding=np.zeros((dataCount,1))
    total = np.zeros((dataCount,1))
    realAction=np.zeros((dataCount,1))
    total[0] = capital
    for ic in range(dataCount):
        currPrice = priceVec[ic]
        if ic > 0:
            stockHolding[ic]=stockHolding[ic-1]
        if suggestedAction[ic] == 1:
            if stockHolding[ic]==0:            
                stockHolding[ic]=capital*(1-transFeeRate)/currPrice
                capital=0
                realAction[ic]=1
        elif suggestedAction[ic] == -1:
            if stockHolding[ic]>0:
                capital=stockHolding[ic]*currPrice*(1-transFeeRate)
                stockHolding[ic]=0
                realAction[ic]=-1
        elif suggestedAction[ic] == 0:
            realAction[ic]=0
        else:
            assert False
        total[ic]=capital+stockHolding[ic]*currPrice*(1-transFeeRate)
    returnRate=(total[-1]-capitalOrig)/capitalOrig
    return returnRate
```

```python
def MAfunc(MAsize,priceVec):
    import numpy as np
    import pandas as pd
    dataLen = len(priceVec)
    MA = np.zeros(dataLen)
    for ic in range(dataLen):
        if ic+1< MAsize:
            MA[ic]=np.sum(priceVec[0:ic+1])/(ic+1)
        else:
            MA[ic]=np.sum(priceVec[ic+1-MAsize:ic+1])/MAsize
    return MA
    
```

```python
def MAoption(MAsize,priceVec,alpha,beta):
    import numpy as np
    import pandas as pd
    dataLen = len(priceVec)
    MA=MAfunc(MAsize,priceVec)
    MAactionVec = np.zeros(dataLen)
    for ic in range(dataLen):
        if priceVec[ic] > MA[ic]+alpha:
            MAactionVec[ic] = 1
        elif priceVec[ic] < MA[ic]+beta:
            MAactionVec[ic] = -1
        else:
            MAactionVec[ic] = 0
    
    
    return MAactionVec
```

```python
def MAtolRate(maxsize,alpha,beta):
    MAreturnRate_tol=np.zeros(maxsize)
    for ic in range(maxsize):
        MAactionVec=MAoption(ic,priceVec,alpha,beta)
        MAreturnRate_tol[ic] = profitEstimateOpen(priceVec, transFeeRate, MAactionVec)    
    sns.set_style("whitegrid")
    plt.figure(figsize=(15,8))
    plt.plot(MAreturnRate_tol)
    plt.plot([0,maxsize],[0,0],'r')
    maxrate=[np.argmax(MAreturnRate_tol),max(MAreturnRate_tol)]
    plt.scatter(maxrate[0],maxrate[1],c='r')
    plt.xlabel('MAsize')
    plt.ylabel('return rates(%)')
    print(maxrate)
    
```

```python
x = Input(shape=(5,))
f_1_1 = Dense(200, activation='sigmoid')
f_1_2 = Dense(200, activation='relu')
f_1_3 = Dense(200, activation='sigmoid')

f_2_1 = Dense(100, activation='sigmoid')
f_2_2 = Dense(100, activation='sigmoid')
f_2_3 = Dense(100, activation='sigmoid')
f_2_4 = Dense(100, activation='sigmoid')


f_3_1 = Dense(50, activation='sigmoid')
f_3_2 = Dense(50, activation='sigmoid')
f_3_3 = Dense(50, activation='sigmoid')

f_4 = Dense(3, activation='softmax')
```

```python
h_1_1 = f_1_1(x)
h_1_2 = f_1_2(x)
h_1_3 = f_1_3(x)
h_1_4 = concatenate([h_1_1,h_1_2])


h_2_1 = f_2_1(h_1_4)
h_2_2 = f_2_2(h_1_4)
h_2_3 = f_2_3(h_1_3)
h_2_4 = f_2_4(h_1_3)
h_2_5 = concatenate([h_2_1,h_2_2,h_2_3])
h_2_6 = concatenate([h_2_1,h_2_4])

h_3_1 = f_3_1(h_2_5)
h_3_2 = f_3_2(h_2_5)
h_3_3 = f_3_3(h_2_6)
h_3 = concatenate([h_3_1,h_3_2,h_3_3])

y = f_4(h_3)
```

```python
model_1=Model(x,y)
model_1.compile(loss='mse', optimizer=SGD(lr=0.087), metrics=['accuracy'])
model_1.summary()
```
